```python
import time
import gym
from stable_baselines3 import DQN

def train_dqn(env_name, total_timesteps=10000):
    # Create the environment
    try:
        env = gym.make(env_name)
    except Exception as e:
        print(f"Error creating environment {env_name}: {e}")
        return None, None, None

    # Initialize the DQN model
    model = DQN('MlpPolicy', env, verbose=1)

    # Start timing the training process
    start_time = time.time()

    # Train the model
    model.learn(total_timesteps=total_timesteps)

    # Calculate training time
    training_time = time.time() - start_time

    # Evaluate the trained model
    obs = env.reset()
    done = False
    total_reward = 0
    episodes = 0

    while not done:
        # Check if the observation is a tuple and unpack it
        if isinstance(obs, tuple):
            obs = obs[0]  # Only take the first part of the tuple

        action, _ = model.predict(obs)
        obs, reward, done, info,_ = env.step(action)
        total_reward += reward
        episodes += 1

    return training_time, episodes, total_reward
environments = ['CartPole-v1', 'LunarLander-v2'] 

results = {}
for env_name in environments:
    training_time, episodes, total_reward = train_dqn(env_name)
    if training_time is not None:
        results[env_name] = {
            'Training Time (s)': training_time,
            'Episodes': episodes,
            'Cumulative Reward': total_reward
        }

print(results)
```
